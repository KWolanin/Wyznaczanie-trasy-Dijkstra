# Wyznaczanie najtańszego kursu lotniczego między stolicami europejskimi za pomocą algorytmu Dijkstry
### Katarzyna Wolanin, informatyka II NST

<img src="files/ross-parmly-25230-unsplash.jpg">

## Opis projektu
Projekt ma na celu wyznaczenie natańszej trasy lotniczej pomiędzy wybranymi miastami-stolicami. Obejmuje graf przedstawiający wzajemne położenie lotnisk stolic europejskich (poszczególne wierzchołki noszą nazwy stolic, z wyjątkiem stolicy Malty - tu użyto nazwy wyspy z powodu, że to jej nazwę nosi lotnisko). Graf opiera się na współrzędnych geograficznych, łączy 44 miasta, z których kilka nie posiada lotnisk (np. Berno), oraz składa się z 386 możliwych połączeń lotniczych. Oprócz grafu w skład projektu wchodzi algorytm, który oblicza najkorzystniejszą cenowo trasę pomiędzy dwoma miastami, informuje o cenie, ilości lotów i, w razie potrzeby, wskazuje kwotę połączenia bezpośredniego (o ile jest droższe niż wytyczone przez algorytm). Dla zobrazowania wyniku stworzono dodatkowy graf wskazujący powstałą trasę.<br>
Dla uproszczenia grafu założono, że wszystkie miasta mają tylko po jednym lotnisku.

## Użyte narzędzia i algorytmy

W projekcie użyto algorytmu Dijkstry, opracowanego przez holenderskiego informatyka Edsgera Dijkstrę. Algorytm ten służy do znajdowania najkrótszej ścieżki z pojedynczego źródła w grafie o nieujemnych wagach krawędzi. W poniższym projekcie źródłem domyślnie jest Warszawa lub wybrane przez użytkownika miasto, w zależności od użytej wersji kodu.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/d/d9/Edsger_Wybe_Dijkstra.jpg/800px-Edsger_Wybe_Dijkstra.jpg" style="width: 200px;">

### Pseudokod algorytmu Dijkstry

```Dijkstra(G,w,s):
   dla każdego wierzchołka v w V[G] wykonaj
      d[v] := nieskończoność
      poprzednik[v] := niezdefiniowane
   d[s] := 0
   Q := V
   dopóki Q niepuste wykonaj
      u := Zdejmij_Min(Q)
      dla każdego wierzchołka v – sąsiada u wykonaj
         jeżeli d[v] > d[u] + w(u, v) to
            d[v] := d[u] + w(u, v)
            poprzednik[v] := u

   Wyświetl("Droga wynosi: " + d[v])```

## Dane
W projekcie wykorzystano współrzędne geograficzne stolic europejskich, zaokrąglone do czterech miejsc po przecinku. Dane odnośnie współrzędnych pobrano z portalu [GeoHack](https://tools.wmflabs.org/geohack) 
używanego przez [Wikipedię](https://pl.wikipedia.org/). Dane odnośnie lotów (połączenia) pochodzą z [Flight Connections](https://www.flightconnections.com/). Ceny połączeń to kwoty na dzień 24.04.19, pobrane z portalu [Sky Scanner](Skyhttps://www.skyscanner.net), przy następujących założeniach: jedna osoba podróżująca, lot w jedną stronę, lot klasą ekonomiczną. Ceny są podane w złotych polskich i są obliczone przez ww. portal jako optymalne.

## Wybór najtańszej trasy od Warszawy do wskazanego przez użytkownika miejsca
#### #1 Przykład użycia algorytmu Dijkstry
Poniższe kody zakładają maksymalnie jedną przesiadkę.

````python
import networkx as nx

end = input("Podaj miasto docelowe: ")
try:
    shortest_path_distance = nx.dijkstra_path(G, source='Warszawa', target=end, weight='price')
    if (nx.algorithms.shortest_paths.generic.has_path(G, source='Warszawa', target=end)):
        print("Najtańsza trasa od: Warszawa do: ", end, "to trasa:", shortest_path_distance)
        path_lenght = nx.dijkstra_path_length(G, source='Warszawa', target=end, weight='price')
        print('Liczba kursów do wykonania to: ', path_lenght)
        li = shortest_path_distance
        middle = li[1]
        if path_lenght==1:
            print('Cena połączenia w PLN: ', price['Warszawa', end])
        if path_lenght==2:
            total = price['Warszawa', middle] + price[middle, end]
            print('Cena połączenia w PLN: ', total)
            if ('Warszawa', end) in price:
                direct = price['Warszawa', end]
                print("Połączenie bezpośrednie kosztuje: ", direct)
                saved = direct - total
                print("Oszczędzasz: ", saved)
        if path_lenght > 2:
            print('Więcej niż jedna przesiadka.')
    else:
        print('Nie ma takiego połączenia!')
except nx.NetworkXNoPath:
    print("Nie znaleziono połączenia - być może miasto nie ma lotniska.")
except nx.NodeNotFound:
    print("To miejsce nie ma lotniska!")
````

## Wybór natańszej trasy: opcja z wyborem źródła
#### #2 Przykład użycia algorytmu Dijkstry

````python
import networkx as nx


start = input("Podaj miasto startowe:")
end = input("Podaj miasto docelowe: ")

try:
    shortest_path_distance = nx.dijkstra_path(G, source=start, target=end, weight='price')
    if (nx.algorithms.shortest_paths.generic.has_path(G, source=start, target=end)):
        print("Najtańsza trasa od: ", start, " do: ", end, "to trasa:", shortest_path_distance)
        path_lenght = nx.dijkstra_path_length(G, source=start, target=end, weight='price')
        print('Liczba kursów do wykonania to: ', path_lenght)
        li = shortest_path_distance
        middle = li[1]
        if path_lenght==1:
            try:
                print('Cena połączenia w PLN: ', price[start, end])
            except KeyError:
                print('Cena połączenia w PLN: ', price[end, start])
        if path_lenght==2:
                try:
                    first_route = price[start, middle]
                except KeyError:
                    first_route = price[middle, start]
                try:
                    second_route = price[middle, end]
                except KeyError:
                    second_route = price[end, middle]
                total = first_route + second_route
                print('Cena połączenia w PLN: ', total)
                if (start, end) in price:
                    direct = price[start, end]
                    print("Połączenie bezpośrednie kosztuje: ", direct)
                    saved = direct - total
                    print("Oszczędzasz: ", saved)
        if path_lenght > 2:
            print('Więcej niż jedna przesiadka.')
    else:
        print('Nie ma takiego połączenia!')
except (nx.NetworkXNoPath, nx.NodeNotFound):
    print("Nie znaleziono połączenia - być może miasto nie ma lotniska bądź wpisano miejsce niepoprawnie.")
````

## Graf obrazujący powstałą trasę (dla wersji z wyborem miejsca startowego i docelowego)

````python
if path_lenght==1:
    try:
        x = price[start, end]
        this_route = {(start, end): x}
    except KeyError:
        x = price[end, start]
        this_route = {(end, start):x}
    x = coordinates.get(start)
    y = coordinates.get(end)
    my_dict = {}
    my_dict[start] = x
    my_dict[end] = y
    print("Trasa: ",start,"-",end, ":")
    x = coordinates.get(start)
    y = coordinates.get(end)
if path_lenght==2:
    try:
        x = price[start, middle]
    except KeyError:
        x = price[middle, start] 
    try:
        y = price[middle, end]   
    except KeyError:
        y = price[end, middle]
    this_route = {(start, middle):x, (middle, end):y}
    x = coordinates.get(start)
    y = coordinates.get(middle)
    z = coordinates.get(end)
    my_dict = {}
    my_dict[start] = x
    my_dict[middle] = y
    my_dict[end] = z
    print("Trasa: ",start,"-",end, ":")
    x = coordinates.get(start)
    y = coordinates.get(middle)
    z = coordinates.get(end)
C = nx.Graph()
C.add_nodes_from(my_dict)
C.add_edges_from(this_route)
plt.figure(figsize=(6,6))
nx.draw(C, my_dict, node_shape='o', node_size=50, with_labels=True, node_color='darkgrey', 
        font_color='red', font_weight='bold', style='dotted', edge_color='darkblue', linewidths=1)
nx.draw_networkx_edge_labels(C, my_dict, edge_labels=this_route)
````
